```python
# Plotting multiple figures/graphs on the same chart
# this is done by first creating the figure, fig = plt.figure()
# you then add subplot, e.g. ax1 = fig.add_subplot()
# then you define each plot using ax1.plot(...)
# doing this multiple times does seem to overlay the previous ones, not sure why this is
```

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
fig1 = plt.figure()
ax1 = fig1.add_subplot(2, 2, 1) # the first two values specify the number of subplot cells
ax2 = fig1.add_subplot(2, 2, 2) # the third value specifies which figure is being referred to
ax3 = fig1.add_subplot(2, 2, 3)
ax4 = fig1.add_subplot(2, 2, 4)
```

```python
ax1.plot(np.random.randn(50).cumsum(), 'k--')
ax2.plot(np.random.randn(200).cumsum(), 'k--')
ax3.plot(np.random.randn(500).cumsum(), 'k--')
ax4.plot(np.random.randn(10000).cumsum(), 'k--')
```

```python
np.random.seed(19680801) # apparently we need to initialise the randomness
fig3 = plt.figure()
ax4 = fig3.add_subplot(2,1,1)
ax4.scatter(np.arange(30), np.arange(30) + 3 * np.random.randn(30)) # a little bit of randomness
ax5 = fig3.add_subplot(2,1,2)
ax5.scatter(np.arange(30), np.arange(30) + 50 * np.random.randn(30)) # a lot of randomness
# learned np.arange(N) - return evenly space values within an interval
# learned np.random.randn(N) - return a sample (or samples) from the “standard normal” distribution.
```
