**How do you create a figure?**

```python
# fig is the whole figure, but axes refer to the plot(s) within the figure 
fig = plt.figure()  # an empty figure with no Axes
fig, ax = plt.subplots()  # a figure with a single Axes
# fig, axs = plt.subplots(2, 2)  # a figure with a 2x2 grid of Axes
```

```python
np.linspace(0,100,11)
# np.linspace return evenly spaced numbers over a specified interval
# numpy.linspace(start, stop, num=50,...) 
# btw should probably do steps is desired steps + 1 as 0 is included
x = np.linspace(0, 2, 100) 
```

```python
# two ways to do graphs in matplotlib
# first way is OO
# Explicitly create figures and axes, and call methods on them 
fig, ax = plt.subplots()  # Create a figure and an axes
ax.plot(x, x, label='linear')  # Plot some data on the axes
ax.plot(x, x**2, label='quadratic')
```

```python
# the other option is to rely on pyplot to automatically create and manage the figures and axes, 
# and use pyplot functions for plotting.
plt.plot(x, x, label='linear')  # Plot some data on the (implicit) axes.
plt.plot(x, x**2, label='quadratic')
```

```python
# what is the difference between the methods?
x = np.linspace(0, 10, 100) 

# OO method - you define a figure using one of the figure initialisers, e.g. fig = plt.figure()
fig, ax = plt.subplots()  # Create a figure and an axes 
print(ax.plot(x, x, label='linear')) # Plot the linear function on the created axes
# pyplot method - use plt.plot(..) but don't specify an axis explicitly
print(plt.plot(x, x**2, label='linear'))  # Plot some data on the (implicit) axes, without explicitly creating an axis
```
